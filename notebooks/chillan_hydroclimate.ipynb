```python
# code to analyze Chillán's hydroclimate
# importing libraries 
import hvplot.pandas
import pandas as pd 

#supress warnings 
import warnings 
warnings.filterwarnings("ignore")
```

```python
# for Río Diguillin en San Lorenzo (Atacalo)
monthly_streamflow = "q_m3s_mon.csv"
daily_streamflow = "q_m3s_day.csv"
daily_precip = "precip_cr2met_day.csv"
tdaymax = "tmax_cr2met_day.csv"
tdaymin = "tmin_cr2met_day.csv"
```

```python
# Make into dataframe - starting with temperature 
tdaymin_df = pd.read_csv(tdaymin)
print(tdaymin_df)
# column name for temp is 8130002, date is date
streamflow_df = pd.read_csv(daily_streamflow)
daily_precip_df = pd.read_csv(daily_precip)
```

```python
# Plot temperature data over time 
min_daily_temp_plot = tdaymin_df.hvplot(
    x='date', y='8130002',
    title='Minimum daily temperature [°C] Atacalo',
    xlabel='Date', ylabel='Temperature [°C]')

display(min_daily_temp_plot)
```

```python
# Plot streamflow data over time 
stream_plot = streamflow_df.hvplot(
    x='date', y='8130002',
    title='Daily streamflow [m3s] Atacalo',
    xlabel='Date', ylabel='Streamflow[m3s]')

display(stream_plot)
```

```python
# Plot streamflow data over time 
precip_plot = daily_precip_df.hvplot(
    x='date', y='8130002',
    title='Daily Precipitation Atacalo',
    xlabel='Date', ylabel='Precipitation [mm]')

display(precip_plot)
```
